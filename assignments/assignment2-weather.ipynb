```python
# Assignment 2
# Author: Jake Daly

import pandas as pd

import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/refs/heads/main/assignment/weatherreadings1.csv"

df = pd.read_csv(url)

pd.set_option('display.max_columns', None)

df.head()
```

```python

# convert date or time strings to datetime objects using the pandas.to_datetime function
df['reportEndDateTime'] = pd.to_datetime(df['reportEndDateTime'])

# plot temperature ('dryBulbTemperature_Celsius') over time ('reportEndDateTime)
plt.figure(figsize=(12,6))  

# plotting the temperature
plt.plot(df['reportEndDateTime'], df['dryBulbTemperature_Celsius'], marker='.', markerfacecolor='yellow', markeredgecolor='red', markersize=8, label='Dry Bulb Temperature (C)')

# adding titles and labels
plt.title('Temperature Over Time', fontsize=16)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Temperature (Celsius)', fontsize=14)
plt.grid(True)  # adding gridlines for clarity


# adjust layout and show plot
plt.tight_layout()
plt.show()

# Resources 

# https://www.datacamp.com/tutorial/matplotlib-time-series-line-plot
```
